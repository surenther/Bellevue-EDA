# Import Data

```python
df_movies = pd.read_csv('movies.csv')
df_movies.head(3)
```

# Explore the Data

```python
df_movies.isna().sum()
```

```python
df_movies.sort_values(by = 'Rating',axis= 0, ascending = False).head(5)
```

```python
# Top 5 movies based on Ratings
df_movies[['Title','Rating','Year']].sort_values(by = 'Rating',axis= 0, ascending = False).head(5)
```

```python
#Top 5 directors by mean Ratings
df_movies[['Directors','Rating']].groupby('Directors').mean() \
.sort_values(by = 'Rating',axis = 0, ascending = False).head(5)
```
